```python
import numpy as np
data = np.genfromtxt('/home/sneha/features.train')
data_final=[]
#seperating the rows which has label 1 and 5
for i in range(data.shape[0]):
    if(data[i][0]==1.0 or data[i][0]==5.0):
        data_final.append(data[i])
data_final=np.array(data_final)
np.random.shuffle(data_final)
y_train=data_final[:,0]
x_train=np.delete(data_final,[0],axis=1)
data = np.genfromtxt('/home/sneha/features.test')
data_final=[]
for i in range(data.shape[0]):
    if(data[i][0]==1.0 or data[i][0]==5.0):
        data_final.append(data[i])
data_final=np.array(data_final)
y_test=data_final[:,0]
x_test=np.delete(data_final,[0],axis=1)
```

```python
#importing libraries
#a_part
from sklearn.multiclass import OneVsOneClassifier
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC 
clf=svm.SVC(random_state=0,kernel='linear',gamma='auto')
model=clf.fit(x_train, y_train)
y_pred=model.predict(x_test)

#Testing error
print('Test accuracy ',accuracy_score(y_test, y_pred))

#Num_of_support_vectors
print('support_vectors',model.n_support_)
```

```python
#b part
#linear kernal
print('***** for 50 rows *****')
y_temp=y_train[:50]
x_temp=x_train[:50]
model=svm.SVC(random_state=0,kernel='linear',gamma='auto').fit(x_temp, y_temp)
y_pred=model.predict(x_test)
#Test accuracy
print('Test_accuracy ',accuracy_score(y_test, y_pred))
#Num_of_support_vectors
print('support_vectors',model.n_support_)
print('------------------------------')
print('***** for 100 rows *****')
y_temp=y_train[:100]
x_temp=x_train[:100]
model=svm.SVC(random_state=0,kernel='linear',gamma='auto').fit(x_temp, y_temp)
y_pred=model.predict(x_test)
print('Test_accuracy ',accuracy_score(y_test, y_pred))
#Num_of_support_vectors
print('support_vectors',model.n_support_)
print('------------------------------')
print('***** for 200 rows *****')
y_temp=y_train[:200]
x_temp=x_train[:200]
model=svm.SVC(random_state=0,kernel='linear',gamma='auto').fit(x_temp, y_temp)
y_pred=model.predict(x_test)
print('Test_accuracy ',accuracy_score(y_test, y_pred))
#Num_of_support_vectors
print('support_vectors',model.n_support_)
print('------------------------------')
print('***** for 800 rows *****')
y_temp=y_train[:800]
x_temp=x_train[:800]
model=svm.SVC(random_state=0,kernel='linear',gamma='auto').fit(x_temp, y_temp)
y_pred=model.predict(x_test)
print('Test_accuracy ',accuracy_score(y_test, y_pred))
#Num_of_support_vectors
print('support_vectors',model.n_support_)
print('------------------------------')
```
