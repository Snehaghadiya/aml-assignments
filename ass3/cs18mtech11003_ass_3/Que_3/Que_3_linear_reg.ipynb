# Que_3, c_part ( i, ii, iii )

```python
# i_part
# it will return the weight vector and the input to this function are X,Y,lamda which are in numpy formate
def mylinridgereg(X, Y, lamda):
    X_transpose=np.transpose(X)
    res = np.dot(X_transpose,X)
    b = np.identity(res.shape[0], dtype = float)
    b=b*lamda
    w=res+b
    try:
        inverse = np.linalg.inv(w)
    except np.linalg.LinAlgError:
        pass
        print('matrix is not invertible')
    weight1=np.dot(inverse,X_transpose)
    final_w=np.dot(weight1,Y)
    return final_w
# ii_part
# it will return the y_pred values and the input to this function are X,weights which are in numpy formate
def mylinridgeregeval(X, weights):
    res=np.dot(X,weights)
    return res
# iii_part
# it will return the mean square error value and the input to this function are y_actaul and y_pred which are in numpy formate
def meansquarederr(A, B):
    mse = (np.square(A - B)).mean()
    return mse
    
```

# Que_3, f_part

```python
#list of test size partition
size_test=[0.1,0.15,0.2,0.4]
li=[]
#forming the list of lamda where step size is of 0.1
for i in range(50):
    li.append(round(i*0.1,1))  
mse_train_total=[]
mse_test_total=[]
for k in size_test:
    mse_train=0.0
    mse_test=0.0
    lamda=[]
    li_train_temp=[]
    li_test_temp=[]
    for i in range(25):
        inside_train=[]
        inside_test=[]
        train, test = train_test_split(s, test_size=k,random_state = 0)
        train_temp = train[[3,4,5,6,7,8,9]]
        test_temp =  test[[3,4,5,6,7,8,9]]
        normalized_train=(train_temp-train_temp.mean())/train_temp.std()
        normalized_test=(test_temp-train_temp.mean())/train_temp.std()
        train1_x=pd.concat([train[[0,1,2]],normalized_train],axis=1)
        test1_x=pd.concat([test[[0,1,2]],normalized_test],axis=1)
        train_x = np.array(train1_x)
        train_y = np.array(train.iloc[:,-1])
        test_x = np.array(test1_x)
        test_y = np.array(test.iloc[:,-1])
        for j in li:
            c = mylinridgereg(train_x, train_y, j)
            target_train_pred = mylinridgeregeval(train_x, c)
            target_test_pred = mylinridgeregeval(test_x, c)
            error_test=meansquarederr(test_y,target_test_pred )
            error_train=meansquarederr(train_y,target_train_pred )
            inside_train.append(error_train)
            inside_test.append(error_test)
        li_train_temp.append(inside_train)
        li_test_temp.append(inside_test)
    li_train_temp=np.array(li_train_temp)
    li_test_temp=np.array(li_test_temp)
    li_train_temp=list(np.sum(li_train_temp,axis=0)/25.0000)
    li_test_temp=list(np.sum(li_test_temp,axis=0)/25.0000)
    mse_train_total.append(li_train_temp)
    mse_test_total.append(li_test_temp)

import matplotlib.pyplot as plt 
i=0
for h in size_test:
    print('for partition ',h)
    # x axis values 
    x = li
    # corresponding y axis values 
    y1 = mse_train_total[i]
    y2= mse_test_total[i]
    # plotting the points  
    plt.plot(x, y1)
   
    plt.plot(x,y2)
    plt.legend(['train_error','test_error'],loc="best")
    # naming the x axis 
    plt.xlabel('lamdA') 
    # naming the y axis 
    plt.ylabel('mse') 
    # giving a title to my graph 
    plt.title('mse versus lamda') 
    # function to show the plot 
#     plt.show() 
#     fig= plt.figure()
    plt.savefig('mse_lamda_test_size_'+str(i))
    i=i+1
    plt.show()
        
```
